# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type

#  1a. Urban total rides
urban_rides = pyber_data_df[(pyber_data_df["type"] == "Urban")].count()["ride_id"]
print(f"{urban_rides} total urban rides")

#  1b. Suburban total rides
suburban_rides = pyber_data_df[(pyber_data_df["type"] == "Suburban")].count()["ride_id"]
print(f"{suburban_rides} total suburban rides")

#  1c. Rural total rides
rural_rides = pyber_data_df[(pyber_data_df["type"] == "Rural")].count()["ride_id"]
print(f"{rural_rides} total rural rides")

# (1) as Series
rides_per_city_type = pyber_data_df.groupby("type")["ride_id"].count()
rides_per_city_type
```

```python
# 2. Get the total drivers for each city type

#  2a. Urban total drivers
urban_drivers = city_data_df[(city_data_df["type"] == "Urban")].sum()["driver_count"]
print(f"{urban_drivers} urban total drivers")

#  2b. Suburban total drivers
suburban_drivers = city_data_df[(city_data_df["type"] == "Suburban")].sum()["driver_count"]
print(f"{suburban_drivers} suburban total drivers")

#  1c. Rural total drivers
rural_drivers = city_data_df[(city_data_df["type"] == "Rural")].sum()["driver_count"]
print(f"{rural_drivers} rural total drivers")

# (1) as Series
#drivers_per_city_type = []
#for idx in pyber_data_df["type"]:
#    for cname in pyber_data_df["city"]:
#        #drivers_per_city_type[idx].append(pyber_data_df["driver_count"])
#        print(pyber_data_df["driver_count"])
drivers_per_city_type = (pyber_data_df.groupby(["type", "city"]).mean()).groupby("type")["driver_count"].sum()
drivers_per_city_type
```

```python
#  3. Get the total amount of fares for each city type

#  3a. Urban total fares
urban_fares_amount = pyber_data_df[(pyber_data_df["type"] == "Urban")].sum()["fare"]
print(f"$US {urban_fares_amount:,.2f} urban total fares")

#  3b. Suburban total fares
suburban_fares_amount = pyber_data_df[(pyber_data_df["type"] == "Suburban")].sum()["fare"]
print(f"$US {suburban_fares_amount:,.2f} suburban total fares")

#  3c. Rural total fares
rural_fares_amount = pyber_data_df[(pyber_data_df["type"] == "Rural")].sum()["fare"]
print(f"$US {rural_fares_amount:,.2f} rural total fares")

# (3) as Series
fares_amount_per_city_type = pyber_data_df.groupby("type")["fare"].sum()
#fares_amount_per_city_type = (pyber_data_df.groupby(["type", "ride_id"]).max()).groupby("type").sum()["fare"]
fares_amount_per_city_type
```

```python
#  4. Get the average fare per ride for each city type. 

#  4a. Urban average fare per ride
urban_average_fare_per_ride = urban_fares_amount / float(urban_rides)
print(f"$US {urban_average_fare_per_ride:.2f} urban average fare per ride")

#  4b. Suburban average fare per ride
suburban_average_fare_per_ride = suburban_fares_amount / float(suburban_rides)
print(f"$US {suburban_average_fare_per_ride:.2f} suburban average fare per ride")

#  4c. Rural average fare per ride
rural_average_fare_per_ride = rural_fares_amount / float(rural_rides)
print(f"$US {rural_average_fare_per_ride:.2f} rural average fare per ride")

# (4) as Series
average_fare_per_ride_per_city_type = fares_amount_per_city_type / rides_per_city_type
##average_fare_per_ride_per_city_type = ( (pyber_data_df.groupby(["type", "ride_id"]).max()).groupby("type").sum()["fare"]
##                                       / (pyber_data_df.groupby("type").count()["ride_id"]))
average_fare_per_ride_per_city_type
```

```python
# 5. Get the average fare per driver for each city type. 

# 5a. Urban average fare per driver
urban_average_fare_per_driver = urban_fares_amount / float(urban_drivers)
print(f"$US {urban_average_fare_per_driver:.2f} urban average fare per driver")

# 5b. Suburban average fare per driver
suburban_average_fare_per_driver = suburban_fares_amount / float(suburban_drivers)
print(f"$US {suburban_average_fare_per_driver:.2f} suburban average fare per driver")

# 5c. Rural average fare per driver
rural_average_fare_per_driver = rural_fares_amount / float(rural_drivers)
print(f"$US {rural_average_fare_per_driver:.2f} rural average fare per driver")

# (5) as Series
average_fare_per_driver_per_city_type = fares_amount_per_city_type / drivers_per_city_type
##average_fare_per_driver_per_city_type = ( (pyber_data_df.groupby(["type", "ride_id"]).max()).groupby("type").sum()["fare"]
##                                 / ((pyber_data_df.groupby(["type", "city"]).mean()).groupby("type").sum()["driver_count"]))
average_fare_per_driver_per_city_type
```

```python
#  6. Create a PyBer summary DataFrame. 

pyber_summary_df = pd.DataFrame(
        {"Total Rides": rides_per_city_type,
         "Total Drivers": drivers_per_city_type,
         "Total Fares": fares_amount_per_city_type,
         "Average Fare per Ride": average_fare_per_ride_per_city_type,
         "Average Fare per Driver": average_fare_per_driver_per_city_type})
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.
pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].map("{:,}".format)
pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].map("{:,.0f}".format)
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:.2f}".format)
pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:.2f}".format)

pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.

pyber_d2_df = pyber_data_df.groupby(["type", "date"])[["fare"]].sum()
pyber_d2_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()

pyber_d2_df = pyber_d2_df.reset_index()

pyber_d2_df
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 

pyber_type_pivot = pyber_d2_df.pivot(index="date", columns="type", values="fare")
pyber_type_pivot.head(10)
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.

rural_fare_series = pyber_type_pivot.loc["2019-01-01":"2019-04-29","Rural"]
suburban_fare_series = pyber_type_pivot.loc["2019-01-01":"2019-04-29","Suburban"]
urban_fare_series = pyber_type_pivot.loc["2019-01-01":"2019-04-29","Urban"]

#print(pyber_type_pivot.loc["2019-01-01":"2019-02-01",:])

pyber_d2new_df = pd.DataFrame(
###             {"Rural": pyber_type_pivot[pyber_type_pivot.loc["date"=="2019-01-01":"2019-04-29",:]],
             {"Rural": rural_fare_series,
              "Suburban": suburban_fare_series,
              "Urban": urban_fare_series})

###pyber_d2new_df = pyber_type_pivot.loc["2019-01-01":"2019-04-29",:]
pyber_d2new_df
```

```python
   # test 5

pyber_d2new_df = pyber_type_pivot.loc["2019-01-01":"2019-04-28",:]
pyber_d2new_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)

pyber_d2new_df.index = pd.to_datetime(pyber_d2new_df.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()

pyber_d2new_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.

#pyber_d2new_df["week_starting"] = pd.date_range("2019-01-01", "2019-04-28", freq="W")

pyber_weekly_fares_df = pyber_d2new_df.resample("W").sum()
pyber_weekly_fares_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

plt.subplots(figsize=(18,6))
ax=plt.gca()

#pyber_weekly_fares_df.plot(kind="line", x="date", y="Rural", ax=ax)
#pyber_weekly_fares_df.plot(kind="line", x="date", y="Suburban", ax=ax)
#pyber_weekly_fares_df.plot(kind="line", x="date", y="Urban", ax=ax)

pyber_weekly_fares_df.plot(y="Rural", ax=ax)
pyber_weekly_fares_df.plot(y="Suburban", ax=ax)
pyber_weekly_fares_df.plot(y="Urban", ax=ax)
plt.title("Total Fare by City Type")
plt.ylabel("Fare ($USD)")
plt.xlabel("")
#lgnd = plt.legend(title="type")
plt.legend(title="type")

plt.savefig("analysis/Pyber_fare_summary.png")
plt.show()
```
